```python
questions_file = "questions.txt"
result_file = "result_data.txt"
keys_file = "keys.txt"

def add_to_result(s):
    with open(result_file, "a", encoding="utf-8" ) as fid:
        fid.write(s+'\n')
        
        
        
# биография

import datetime
time = datetime.datetime.now().isoformat(timespec='minutes', sep=' ')
with open (result_file, "w", encoding="utf-8") as fid:
    fid.write(time+'\n')
        
add_to_result("ФИО:"+input("Добрый день, уважаемый участник! Введите ФИО:" ))
add_to_result("Дата рождения"+input("Введите дату рождения:")+'\n')


# шапка

print("""
Сейчас Вам будут предложены утверждения, касающиеся Вашего характера.
Если Вы согласны с утверждением - введите цифру 1, если нет - цифру 0.
Для перехода к следующему вопросу - нажмите Enter.
Над вопросами долго не думайте, правильных и неправильных ответов нет\n""")


# список вопросов
with open(questions_file, "r", encoding="utf-8") as fid:
    questions = fid.readlines() 
    
    

# цикл для ответов
answers = []
n_questions = len(questions)
i = 0
while i < n_questions:
    print()
    print(questions[i])
    answ = int(input("Ответ 1(да) или 0(нет):"))
    if answ == 1 or answ == 0:
        answers.append(answ)
        i += 1
    else: 
        print("Введите ответ 1 или 0")
        i = i

    
# запись ответов в файл
for n, ans in enumerate(answers):
    s = str(n+1)+" : "+str(ans)
    add_to_result(s)
    


    
factors = {}
multiplications = {}

with open(keys_file, "r", encoding="utf-8") as fid:
    lines = fid.readlines()
    
for line in lines:
    factors[line.split()[2]] = list(map(int, line.split()[4].split(",")))
    multiplications [line.split()[2]] = int(line.split()[3][1])
    
    

    
scales = {}

for factorname, keys_quest in factors.items():
    for n in range(len(keys_quest)):   # отнимаем единицу или прибавляем в зависимости от того обратный вопрос или прямой
        if keys_quest[n] >= 0:
            keys_quest[n] -= 1
        else:
             keys_quest[n] += 1



# подсчет результатов
for factorname, keys_quest in factors.items():
    factor_value = 0
    for n in range(len(keys_quest)):
        if keys_quest[n] >= 0:
            factor_value += answers[keys_quest[n]]
        else:
            factor_value -= answers[abs(keys_quest[n])]

    scales [factorname] = factor_value * multiplications[factorname]
    add_to_result(factorname+":"+str(scales[factorname]))
    
    


print("____" * 20)
print(" " * 30 + "РЕЗУЛЬТАТЫ:")

accentuations = []
for factorname, value in scales.items():
    if value > 12:
        accentuations.append(factorname)
           
res_for_print = ', '.join(accentuations)

if len(accentuations) > 0:
    print(f"У вас выялены акцентуации характера по следующим типам: {res_for_print}")
else:
    print("У вас не выявлены акцентуации характера")
    
            
     
```
